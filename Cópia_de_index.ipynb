````python
# example of updating prices using a for loop and if/else

produtos = [
    {"nome": "Produto A", "preco": 100},
    {"nome": "Produto B", "preco": 250},
    {"nome": "Produto C", "preco": 50},
]
print("PreÃ§os originais:")
for produto in produtos:
    print(f"{produto['nome']}: R${produto['preco']:.2f}")

#apply price updates
for produto in produtos:
    if produto['preco'] < 100:
        produto['preco'] *= 1.15 # increase price by 15% for products under R$ 100
    elif produto['preco'] >= 100 and produto['preco'] <= 200:
        produto['preco'] *= 1.10 # increase price by 10% for products between R$ 100 and R$ 200
    else:
        produto['preco'] *= 1.05 # increase price by 5% for products over R$ 200

print("\nPreÃ§os atualizados:")
for produto in produtos:
    print(f"{produto['nome']}: R${produto['preco']:.2f}")
````

#### CÃ¡lculo da MÃ©dia
<hr style="border: 1px solid rgba(242, 248, 241, 1); width: 100%; margin: 5px 0;">
A fÃ³rmula da mÃ©dia Ã© usada para calcular o valor mÃ©dio de um conjunto de dados. A fÃ³rmula Ã© dada por:
$$
\text{MÃ©dia} = \frac{\sum_{i=1}^{n} x_i}{n}
$$
Onde:
- $( x_i )$ representa cada valor do conjunto de dados.
- $( n )$ Ã© o nÃºmero total de elementos.
### Exemplo:
Se tivermos os seguintes nÃºmeros: 2, 4, 6, 8, 10, a mÃ©dia serÃ¡:
$$
\text{MÃ©dia} = \frac{2 + 4 + 6 + 8 + 10}{5} = \frac{30}{5}Â =Â 6
$$

### **Projeto de AnÃ¡lise de Dados: OtimizaÃ§Ã£o de Rentabilidade de Produtos**

[cite\_start]**Empresa:** T\&P Revestimentos LTDA [cite: 2]
[cite\_start]**Fonte de Dados:** RelatÃ³rio de Consultoria ContÃ¡bil e Financeira (Janeiro a Junho/2025) [cite: 1, 3]

#### **1. Problema a Ser Solucionado**

[cite\_start]O relatÃ³rio financeiro da T\&P Revestimentos aponta um problema central: apesar de um faturamento bruto expressivo de **R$ 1.761.954,00**, a empresa encerrou o semestre com um **prejuÃ­zo operacional de R$ 286.506,37**[cite: 5, 105]. [cite\_start]A anÃ¡lise preliminar indica que a margem bruta Ã© muito baixa para o volume faturado[cite: 86, 90].

O problema a ser solucionado Ã©: **Identificar quais produtos possuem a menor rentabilidade e mais impactam negativamente o resultado da empresa, a fim de direcionar aÃ§Ãµes estratÃ©gicas para otimizar o portfÃ³lio de vendas.**

#### **2. HipÃ³teses para Orientar a AnÃ¡lise**

Para guiar a investigaÃ§Ã£o, foram elaboradas as seguintes hipÃ³teses:

  * **HipÃ³tese 1:** Os produtos com maior volume de receita (como Piso VinÃ­lico) nÃ£o sÃ£o necessariamente os mais lucrativos e podem possuir margens de lucro apertadas, contribuindo para o baixo resultado geral.
  * **HipÃ³tese 2:** Existe uma concentraÃ§Ã£o de custos de material nos produtos mais vendidos, o que pressiona a margem de lucro bruto para baixo.
  * **HipÃ³tese 3:** Produtos de menor expressÃ£o em receita podem ter margens de lucro maiores, representando oportunidades ocultas que poderiam ser mais bem exploradas.

#### **3. Desenvolvimento da SoluÃ§Ã£o com AnÃ¡lise de Dados**

Para validar essas hipÃ³teses, vamos utilizar a biblioteca **Pandas** do Python para estruturar os dados do relatÃ³rio, calcular o lucro bruto e a margem de cada produto.

##### **CÃ³digo em Python (Comentado)**

```python
# Passo 1: Importar a biblioteca Pandas, essencial para a manipulaÃ§Ã£o de dados em tabelas.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Passo 2: Criar um DataFrame com os dados de RECEITA por produto, extraÃ­dos do relatÃ³rio.
# Um DataFrame Ã© uma tabela de dados, a principal estrutura do Pandas.
dados_receita = {
    'Produto': [
        'Piso VinÃ­lico', 'Carpete', 'Grama SintÃ©tica', 'Venda Havan', 'Piso VinÃ­lico SPC',
        'Venda Loja', 'ServiÃ§os', 'Personalizado', 'Massa e Insumos', 'RodapÃ©'
    ],
    'Receita_Bruta': [
        524678.38, 359232.76, 283697.32, 199161.04, 105251.23,
        53141.88, 43214.80, 25019.32, 19872.65, 14495.48
    ]
}
df_receitas = pd.DataFrame(dados_receita)

# Passo 3: Criar um DataFrame com os CUSTOS de compra de material por produto.
# Esses sÃ£o os custos diretos para a venda de cada item (Custo da Mercadoria Vendida).
dados_custos = {
    'Produto': [
        'Piso VinÃ­lico', 'Carpete', 'Massa e Insumos', 'Grama SintÃ©tica',
        'Personalizado', 'RodapÃ©'
    ],
    'Custo_Material': [
        225994.93, 149632.54, 101517.00, 89469.43,
        9824.73, 8813.54
    ]
}
df_custos = pd.DataFrame(dados_custos)

# Passo 4: Unir (merge) os dois DataFrames em um Ãºnico para centralizar a anÃ¡lise.
# A uniÃ£o Ã© feita com base na coluna 'Produto'. Produtos sem custo de material correspondente
# ficarÃ£o com o valor NaN (Not a Number).
df_analise = pd.merge(df_receitas, df_custos, on='Produto', how='left')

# Passo 5: Calcular as mÃ©tricas de rentabilidade.
# O Lucro Bruto Ã© a diferenÃ§a entre o que foi vendido e o que custou para comprar o produto.
df_analise['Lucro_Bruto'] = df_analise['Receita_Bruta'] - df_analise['Custo_Material']

# A Margem de Lucro Bruto percentual mostra a rentabilidade de cada real vendido.
df_analise['Margem_Lucro_%'] = (df_analise['Lucro_Bruto'] / df_analise['Receita_Bruta']) * 100

# Passo 6: Ordenar o DataFrame para facilitar a visualizaÃ§Ã£o dos resultados.
# Primeiro, ordenamos pela margem para ver os mais e menos rentÃ¡veis.
df_analise_ordenada = df_analise.sort_values(by='Margem_Lucro_%', ascending=False)

# Exibir a tabela final com os resultados. O .round(2) formata para 2 casas decimais.
print("--- AnÃ¡lise de Rentabilidade por Produto ---")
print(df_analise_ordenada.round(2))

# Passo 7: VisualizaÃ§Ã£o dos Dados para a apresentaÃ§Ã£o.
# Um grÃ¡fico de barras ajuda a comparar visualmente o lucro bruto de cada produto.
plt.figure(figsize=(12, 8)) # Define o tamanho do grÃ¡fico.
```

A anÃ¡lise de dados confirma que a baixa rentabilidade da T\&P Revestimentos estÃ¡ diretamente ligada a problemas na composiÃ§Ã£o de seu portfÃ³lio de vendas. A performance de "Massa e Insumos" Ã© insustentÃ¡vel e precisa de uma revisÃ£o urgente.

````python
# Passo 1: Importar a biblioteca Pandas, essencial para a manipulaÃ§Ã£o de dados em tabelas.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Passo 2: Criar um DataFrame com os dados de RECEITA por produto, extraÃ­dos do relatÃ³rio.
# Um DataFrame Ã© uma tabela de dados, a principal estrutura do Pandas.
dados_receita = {
    'Produto': [
        'Piso VinÃ­lico', 'Carpete', 'Grama SintÃ©tica', 'Venda Havan', 'Piso VinÃ­lico SPC',
        'Venda Loja', 'ServiÃ§os', 'Personalizado', 'Massa e Insumos', 'RodapÃ©'
    ],
    'Receita_Bruta': [
        524678.38, 359232.76, 283697.32, 199161.04, 105251.23,
        53141.88, 43214.80, 25019.32, 19872.65, 14495.48
    ]
}
df_receitas = pd.DataFrame(dados_receita)

# Passo 3: Criar um DataFrame com os CUSTOS de compra de material por produto.
# Esses sÃ£o os custos diretos para a venda de cada item (Custo da Mercadoria Vendida).
dados_custos = {
    'Produto': [
        'Piso VinÃ­lico', 'Carpete', 'Massa e Insumos', 'Grama SintÃ©tica',
        'Personalizado', 'RodapÃ©'
    ],
    'Custo_Material': [
        225994.93, 149632.54, 101517.00, 89469.43,
        9824.73, 8813.54
    ]
}
df_custos = pd.DataFrame(dados_custos)

# Passo 4: Unir (merge) os dois DataFrames em um Ãºnico para centralizar a anÃ¡lise.
# A uniÃ£o Ã© feita com base na coluna 'Produto'. Produtos sem custo de material correspondente
# ficarÃ£o com o valor NaN (Not a Number).
df_analise = pd.merge(df_receitas, df_custos, on='Produto', how='left')

# Passo 5: Calcular as mÃ©tricas de rentabilidade.
# O Lucro Bruto Ã© a diferenÃ§a entre o que foi vendido e o que custou para comprar o produto.
df_analise['Lucro_Bruto'] = df_analise['Receita_Bruta'] - df_analise['Custo_Material']

# A Margem de Lucro Bruto percentual mostra a rentabilidade de cada real vendido.
df_analise['Margem_Lucro_%'] = (df_analise['Lucro_Bruto'] / df_analise['Receita_Bruta']) * 100

# Passo 6: Ordenar o DataFrame para facilitar a visualizaÃ§Ã£o dos resultados.
# Primeiro, ordenamos pela margem para ver os mais e menos rentÃ¡veis.
df_analise_ordenada = df_analise.sort_values(by='Margem_Lucro_%', ascending=False)

# Replace NaN values with 0
df_analise_ordenada[['Lucro_Bruto', 'Margem_Lucro_%']] = df_analise_ordenada[['Lucro_Bruto', 'Margem_Lucro_%']].fillna(0)


# Exibir a tabela final com os resultados. O .round(2) formata para 2 casas decimais.
print("--- AnÃ¡lise de Rentabilidade por Produto ---")
print(df_analise_ordenada.round(2))

# Passo 7: VisualizaÃ§Ã£o dos Dados para a apresentaÃ§Ã£o.
# Um grÃ¡fico de barras ajuda a comparar visualmente o lucro bruto de cada produto.
plt.figure(figsize=(12, 8)) # Define o tamanho do grÃ¡fico.
sns.barplot(x='Lucro_Bruto', y='Produto', data=df_analise_ordenada, palette='viridis')
plt.title('Lucro Bruto por Linha de Produto (1Âº Semestre 2025)', fontsize=16)
plt.xlabel('Lucro Bruto (R$)', fontsize=12)
plt.ylabel('Produto', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
````
