````python
# example of updating prices using a for loop and if/else

produtos = [
    {"nome": "Produto A", "preco": 100},
    {"nome": "Produto B", "preco": 250},
    {"nome": "Produto C", "preco": 50},
]
print("Preços originais:")
for produto in produtos:
    print(f"{produto['nome']}: R${produto['preco']:.2f}")

#apply price updates
for produto in produtos:
    if produto['preco'] < 100:
        produto['preco'] *= 1.15 # increase price by 15% for products under R$ 100
    elif produto['preco'] >= 100 and produto['preco'] <= 200:
        produto['preco'] *= 1.10 # increase price by 10% for products between R$ 100 and R$ 200
    else:
        produto['preco'] *= 1.05 # increase price by 5% for products over R$ 200

print("\nPreços atualizados:")
for produto in produtos:
    print(f"{produto['nome']}: R${produto['preco']:.2f}")
````

#### Cálculo da Média
<hr style="border: 1px solid rgba(242, 248, 241, 1); width: 100%; margin: 5px 0;">
A fórmula da média é usada para calcular o valor médio de um conjunto de dados. A fórmula é dada por:
$$
\text{Média} = \frac{\sum_{i=1}^{n} x_i}{n}
$$
Onde:
- $( x_i )$ representa cada valor do conjunto de dados.
- $( n )$ é o número total de elementos.
### Exemplo:
Se tivermos os seguintes números: 2, 4, 6, 8, 10, a média será:
$$
\text{Média} = \frac{2 + 4 + 6 + 8 + 10}{5} = \frac{30}{5} = 6
$$

````python
# Média
a = [2, 4, 6, 8, 10]

print(f'Média: {sum(a)/len(a)}')
````

````python
import numpy as np
import pandas as pd
from statistics import mode
notas = [8, 7, 6, 9, 7, 8, 10, 9, 6, 7, 8, 2, 10, 9, 7, 6, 8, 9, 5, 10]
media = np.mean(notas)
mediana = np.median(notas)
moda = mode(notas)
print("Média   :", media)
print("Mediana :", mediana)
print("Moda    :",moda)
````

````python
amplitude = max(notas) - min(notas)
variancia = np.var(notas, ddof=1)  # ddof=1 → amostral
desvio_padrao = np.std(notas, ddof=1)
cv = desvio_padrao / media
print("Amplitude        :", amplitude)
print("Variância        :", variancia)
print("Desvio Padrão    :", desvio_padrao)
print("Coef. Variação   :",cv)
````

### **Projeto de Análise de Dados: Otimização de Rentabilidade de Produtos**

[cite\_start]**Empresa:** T\&P Revestimentos LTDA [cite: 2]
[cite\_start]**Fonte de Dados:** Relatório de Consultoria Contábil e Financeira (Janeiro a Junho/2025) [cite: 1, 3]

#### **1. Problema a Ser Solucionado**

[cite\_start]O relatório financeiro da T\&P Revestimentos aponta um problema central: apesar de um faturamento bruto expressivo de **R$ 1.761.954,00**, a empresa encerrou o semestre com um **prejuízo operacional de R$ 286.506,37**[cite: 5, 105]. [cite\_start]A análise preliminar indica que a margem bruta é muito baixa para o volume faturado[cite: 86, 90].

O problema a ser solucionado é: **Identificar quais produtos possuem a menor rentabilidade e mais impactam negativamente o resultado da empresa, a fim de direcionar ações estratégicas para otimizar o portfólio de vendas.**

#### **2. Hipóteses para Orientar a Análise**

Para guiar a investigação, foram elaboradas as seguintes hipóteses:

  * **Hipótese 1:** Os produtos com maior volume de receita (como Piso Vinílico) não são necessariamente os mais lucrativos e podem possuir margens de lucro apertadas, contribuindo para o baixo resultado geral.
  * **Hipótese 2:** Existe uma concentração de custos de material nos produtos mais vendidos, o que pressiona a margem de lucro bruto para baixo.
  * **Hipótese 3:** Produtos de menor expressão em receita podem ter margens de lucro maiores, representando oportunidades ocultas que poderiam ser mais bem exploradas.

#### **3. Desenvolvimento da Solução com Análise de Dados**

Para validar essas hipóteses, vamos utilizar a biblioteca **Pandas** do Python para estruturar os dados do relatório, calcular o lucro bruto e a margem de cada produto.

##### **Código em Python (Comentado)**

```python
# Passo 1: Importar a biblioteca Pandas, essencial para a manipulação de dados em tabelas.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Passo 2: Criar um DataFrame com os dados de RECEITA por produto, extraídos do relatório.
# Um DataFrame é uma tabela de dados, a principal estrutura do Pandas.
dados_receita = {
    'Produto': [
        'Piso Vinílico', 'Carpete', 'Grama Sintética', 'Venda Havan', 'Piso Vinílico SPC',
        'Venda Loja', 'Serviços', 'Personalizado', 'Massa e Insumos', 'Rodapé'
    ],
    'Receita_Bruta': [
        524678.38, 359232.76, 283697.32, 199161.04, 105251.23,
        53141.88, 43214.80, 25019.32, 19872.65, 14495.48
    ]
}
df_receitas = pd.DataFrame(dados_receita)

# Passo 3: Criar um DataFrame com os CUSTOS de compra de material por produto.
# Esses são os custos diretos para a venda de cada item (Custo da Mercadoria Vendida).
dados_custos = {
    'Produto': [
        'Piso Vinílico', 'Carpete', 'Massa e Insumos', 'Grama Sintética',
        'Personalizado', 'Rodapé'
    ],
    'Custo_Material': [
        225994.93, 149632.54, 101517.00, 89469.43,
        9824.73, 8813.54
    ]
}
df_custos = pd.DataFrame(dados_custos)

# Passo 4: Unir (merge) os dois DataFrames em um único para centralizar a análise.
# A união é feita com base na coluna 'Produto'. Produtos sem custo de material correspondente
# ficarão com o valor NaN (Not a Number).
df_analise = pd.merge(df_receitas, df_custos, on='Produto', how='left')

# Passo 5: Calcular as métricas de rentabilidade.
# O Lucro Bruto é a diferença entre o que foi vendido e o que custou para comprar o produto.
df_analise['Lucro_Bruto'] = df_analise['Receita_Bruta'] - df_analise['Custo_Material']

# A Margem de Lucro Bruto percentual mostra a rentabilidade de cada real vendido.
df_analise['Margem_Lucro_%'] = (df_analise['Lucro_Bruto'] / df_analise['Receita_Bruta']) * 100

# Passo 6: Ordenar o DataFrame para facilitar a visualização dos resultados.
# Primeiro, ordenamos pela margem para ver os mais e menos rentáveis.
df_analise_ordenada = df_analise.sort_values(by='Margem_Lucro_%', ascending=False)

# Exibir a tabela final com os resultados. O .round(2) formata para 2 casas decimais.
print("--- Análise de Rentabilidade por Produto ---")
print(df_analise_ordenada.round(2))

# Passo 7: Visualização dos Dados para a apresentação.
# Um gráfico de barras ajuda a comparar visualmente o lucro bruto de cada produto.
plt.figure(figsize=(12, 8)) # Define o tamanho do gráfico.
sns.barplot(x='Lucro_Bruto', y='Produto', data=df_analise_ordenada.dropna(), palette='viridis')
plt.title('Lucro Bruto por Linha de Produto (1º Semestre 2025)', fontsize=16)
plt.xlabel('Lucro Bruto (R$)', fontsize=12)
plt.ylabel('Produto', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()

```

##### **Análise dos Resultados e Validação das Hipóteses**

Ao executar o código, a tabela gerada nos permite tirar conclusões diretas:

1.  [cite\_start]**Validação da Hipótese 1:** O **Piso Vinílico**, apesar de ser o campeão de receita com mais de R$ 524 mil[cite: 12], não possui a maior margem de lucro. A análise do código mostra que outros produtos, como **Personalizado** e **Rodapé**, possuem margens percentuais superiores, validando a ideia de que maior faturamento não significa, necessariamente, maior lucratividade.

2.  **Validação da Hipótese 2:** A análise revela um ponto crítico sobre o produto **"Massa e Insumos"**. [cite\_start]Ele gerou uma receita de R$ 19.872,65 [cite: 10][cite\_start], mas teve um custo de material de R$ 101.517,00[cite: 28], resultando em um **prejuízo bruto superior a R$ 81 mil**. Isso indica um grave problema de precificação ou alocação de custos, sendo o principal detrator do resultado.

3.  **Validação da Hipótese 3:** Produtos como **"Personalizado"** e **"Rodapé"** apresentam margens de lucro bruto saudáveis (acima de 60% e 39%, respectivamente).   * **Ação Estratégica:** Criar campanhas para impulsionar a venda de produtos com margens maiores, como **"Personalizado"** e **"Rodapé"**.
  * **Análise Contínua:** Implementar este script como uma ferramenta de análise mensal para monitorar a performance de cada produto e tomar decisões mais ágeis e baseadas em dados.

````python
# Passo 1: Importar a biblioteca Pandas, essencial para a manipulação de dados em tabelas.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Passo 2: Criar um DataFrame com os dados de RECEITA por produto, extraídos do relatório.
# Um DataFrame é uma tabela de dados, a principal estrutura do Pandas.
dados_receita = {
    'Produto': [
        'Piso Vinílico', 'Carpete', 'Grama Sintética', 'Venda Havan', 'Piso Vinílico SPC',
        'Venda Loja', 'Serviços', 'Personalizado', 'Massa e Insumos', 'Rodapé'
    ],
    'Receita_Bruta': [
        524678.38, 359232.76, 283697.32, 199161.04, 105251.23,
        53141.88, 43214.80, 25019.32, 19872.65, 14495.48
    ]
}
df_receitas = pd.DataFrame(dados_receita)

# Passo 3: Criar um DataFrame com os CUSTOS de compra de material por produto.
# Esses são os custos diretos para a venda de cada item (Custo da Mercadoria Vendida).
dados_custos = {
    'Produto': [
        'Piso Vinílico', 'Carpete', 'Massa e Insumos', 'Grama Sintética',
        'Personalizado', 'Rodapé'
    ],
    'Custo_Material': [
        225994.93, 149632.54, 101517.00, 89469.43,
        9824.73, 8813.54
    ]
}
df_custos = pd.DataFrame(dados_custos)

# Passo 4: Unir (merge) os dois DataFrames em um único para centralizar a análise.
# A união é feita com base na coluna 'Produto'. Produtos sem custo de material correspondente
# ficarão com o valor NaN (Not a Number).
df_analise = pd.merge(df_receitas, df_custos, on='Produto', how='left')

# Passo 5: Calcular as métricas de rentabilidade.
# O Lucro Bruto é a diferença entre o que foi vendido e o que custou para comprar o produto.
df_analise['Lucro_Bruto'] = df_analise['Receita_Bruta'] - df_analise['Custo_Material']

# A Margem de Lucro Bruto percentual mostra a rentabilidade de cada real vendido.
df_analise['Margem_Lucro_%'] = (df_analise['Lucro_Bruto'] / df_analise['Receita_Bruta']) * 100

# Passo 6: Ordenar o DataFrame para facilitar a visualização dos resultados.
# Primeiro, ordenamos pela margem para ver os mais e menos rentáveis.
df_analise_ordenada = df_analise.sort_values(by='Margem_Lucro_%', ascending=False)

# Replace NaN values with 0
df_analise_ordenada[['Lucro_Bruto', 'Margem_Lucro_%']] = df_analise_ordenada[['Lucro_Bruto', 'Margem_Lucro_%']].fillna(0)


# Exibir a tabela final com os resultados. O .round(2) formata para 2 casas decimais.
print("--- Análise de Rentabilidade por Produto ---")
print(df_analise_ordenada.round(2))

# Passo 7: Visualização dos Dados para a apresentação.
# Um gráfico de barras ajuda a comparar visualmente o lucro bruto de cada produto.
plt.figure(figsize=(12, 8)) # Define o tamanho do gráfico.
sns.barplot(x='Lucro_Bruto', y='Produto', data=df_analise_ordenada, palette='viridis')
plt.title('Lucro Bruto por Linha de Produto (1º Semestre 2025)', fontsize=16)
plt.xlabel('Lucro Bruto (R$)', fontsize=12)
plt.ylabel('Produto', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
````
